Exploratory Data Analysis (EDA) Summary

Overview:
In this Exploratory Data Analysis (EDA), I analyzed a dataset for ELA and Math grades for several years during COVID-19. The dataset includes information such as DBN (District Borough Number), School Name, Category, Grade Metrics, and Year.

Key Findings:


```python
import os
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

Key Variables:

'Borough': The district where the students live and go to school in.

'Category': The type of student being assessed

'Grade': The School grade which test was administered: All Grades, 3-8 grades

'Year': The year that the NYS standardized exam was taken. 

'Category' : Reflects the set of students which were tested

'Number Tested' : The total number of students test

'Mean Scale Score' : The Average of total students tested

'# Level 1' : Number of students who scored in Level 1 range

'% Level 1' : Percentage of students who scored in Level 1 range

'# Level 2': Number of students who scored in Level 2 range

'% Level 2': Percentage of students who scored in Level 2 range

'# Level 3': Number of students who scored in Level 3 range

'% Level 3': Percentage of students who scored in Level 3 range

'# Level 4': Number of students who scored in Level 4 range

'% Level 4': Percentage of students who scored in Level 4 range

'# Level 3+4': Number of students who scored in Level 3+4 range

'% Level 3+4': Percentage of students who scored in Level 3+4 range

```python
# Plotting the graph
plt.figure(figsize=(15, 10))
plt.plot(grades_df.index, grades_df['% Level 1'], marker='o', label='Level 1')
plt.plot(grades_df.index, grades_df['% Level 2'], marker='o', label='Level 2')
plt.plot(grades_df, grades_df['% Level 3'], marker='o', label='Level 3')
plt.plot(grades_df.index, grades_df['% Level 4'], marker='o', label='Level 4')

plt.title('Percentage of Students at Four Distinct Achievement Levels')
plt.xlabel('Grade')
plt.ylabel('Percentage of Students')
plt.legend(title='Different Achievement Levels')
plt.grid(True)

plt.show()
```

```python
#plotting for students who got Grade 4 Level Achievement
plt.figure(figsize=(20, 12))
plt.plot(grades_df['% Level 4'], marker='o', linestyle='-', label='Level 4')

plt.title('Percentage of Students at Level 4 Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend(title='Achievement Level')
plt.grid(True)

plt.show()
```
